# SigSys Adaptive Filter Exploration

```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy as np
import math
import pandas
from scipy import signal, fftpack
import dct
import dft
import autocorr

PI2 = 2 * math.pi

%matplotlib inline
```

```python
def lms_filt(wave,d_wave,step=1,P=10):
    """
    Takes an input wave, and the desired signal
    step for gradient descent
    P for number of taps
    """
    
    #make first arrays and matrices
    inp = wave.ys
    des = d_wave.ys
    K = len(inp)-P+1
    out = np.zeros(K)
    err = np.zeros(K)
    coeff = np.zeros(P)
    weights = np.zeros((K,P))
    
    #apply filter and algorithm
    for k in range(K):
        x = np.flipud(inp[k:k+P])
        out[k] = np.dot(x, coeff)
        err[k] = inp[k+P-1] - out[k]
        coeff = coeff + step * x * err[k]
        out[k] = np.dot(x, coeff)
        weights[k] = coeff
    
    return out, err, weights
```

```python
wave = thinkdsp.Sinusoid(freq=440, amp=0.5, offset=0).make_wave(duration=.02)
d_wave = wave
noisy = wave.ys + np.random.uniform(-.1,.1,(len(wave.ys)))
wave = thinkdsp.Wave(noisy, 11025)
wave.plot()
```

```python
step = .02
P = 20
new, err, weights = lms_filt(wave,d_wave,step,P)
new_wave = thinkdsp.Wave(new, 11025)
new_wave.plot()
```
