# SigSys Adaptive Filter Exploration - Implementing Least Mean Squares

```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy as np
import math
import pandas
from scipy import signal, fftpack
import dct
import dft
import autocorr

PI2 = 2 * math.pi

%matplotlib inline
```

We'll start out by defining a function which takes in the desired signal, the output signal, and some parameters for the LMS algorithm. It will then output the resultant error, signal, and weights.

```python
def lms_filt(wave,d_wave,step=1,P=10):
    """
    Takes an input wave, and the desired signal
    step for gradient descent
    P for number of taps
    
    Returns error signal, predicted output, and the weights for the filter
    """
    
    #make first arrays and matrices
    inp = wave.ys
    des = d_wave.ys
    K = len(inp)-P+1
    out = np.zeros(K)
    err = np.zeros(K)
    coeff = np.zeros(P)
    weights = np.zeros((K,P))
    
    #apply filter and algorithm
    for k in range(K):
        x = np.flipud(inp[k:k+P])
        out[k] = np.dot(x, coeff)
        err[k] = inp[k+P-1] - out[k]
        coeff = coeff + step * x * err[k]
        out[k] = np.dot(x, coeff)
        weights[k] = coeff
    
    return out, err, weights
```

Next, we'll generate a signal using Allen Downey's ThinkDSP library. We'll add some noise to it and then see how it looks.

```python
wave = thinkdsp.Sinusoid(freq=440, amp=0.5, offset=0).make_wave(duration=.02)
d_wave = wave #the desired wave
noisy = wave.ys + np.random.uniform(-.1,.1,(len(wave.ys))) #add randomized noise
wave = thinkdsp.Wave(noisy, 11025)
wave.plot()
```

It's time to set some parameters for the LMS. We'll use a step size of .02 for the gradient descent because it doesn't hurt to start small. You'll have rebounding issues if you start to big, as the algorithm will overshoot where it thinks the minima is during gradient descent and rebound through the cost curve out of control. We'll set the number of taps to be fairly high, so as to get a really good approximation.

```python
step = .02
P = 20
new, err, weights = lms_filt(wave,d_wave,step,P)
new_wave = thinkdsp.Wave(new, 11025)
new_wave.plot()
```

As expected, it does a fairly good job at predicting the signal after some time has passed. It overcompensates at the beginning but eventually the signal is better approximated. We can confirm this by plotting the mean squared error overtime.

```python
mswe = np.mean((weights - np.concatenate(([0],[1],np.zeros(18))))**2, axis=1)
thinkplot.Plot(np.diff(mswe))
```

Again, as expected, the error overall is reduced to close to zero. There are small jitters in the graph most likely due to the high number of taps or the size of the step, but for the most part the error is eventually close to zero.
